```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth

scope = ["user-library-read","user-library-modify","playlist-modify-private","playlist-modify-public"]

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope,client_id='.',client_secret='.',redirect_uri='http://localhost:3000'))
```

```python
liked=sp.current_user_saved_tracks(limit=10000)
```

```python
import pandas as pd
data=pd.DataFrame(columns=['track_name','artist_name','album_name','track_id','artist_id','album_id'])
```

```python
for like in liked['items']:
    data.loc[len(data)]=[
    like['track']['name'] ,
    ','.join([artist['name'] for artist in like['track']['artists']]),
    like['track']['album']['name'],
    like['track']['id'],
    ','.join([artist['id'] for artist in like['track']['artists']]),
    like['track']['album']['id']
    ]
```
